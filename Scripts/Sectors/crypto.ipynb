```python
coins = openbb.crypto.disc.losers(interval = '1y').sort_values('Change 1y [%]', ascending=False)
coins.head(10)
```

```python
coins_pos = coins.loc[coins['Change 1y [%]'] > 0 ].reset_index(drop=True)
coins_pos
```

```python
coins_neg = coins.loc[coins['Change 1y [%]'] < 0 ].sort_values('Change 1y [%]', ascending=True).reset_index(drop=True)
coins_neg
```
