```python
from openbb_terminal.sdk import openbb
from datetime import datetime
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
%matplotlib inline
```

```python
from dateutil.relativedelta import relativedelta 

def load_futures(symbol):
    end_date = datetime.now().date()
    start_date = end_date - relativedelta(months= 1)
    df = openbb.futures.historical([symbol],start_date=start_date, end_date=end_date)
    # df = df.loc[start_date : end_date]
    df = df.iloc[[0, -1]]
    return (df.iloc[0]['Adj Close'].round(3), df.iloc[1]['Adj Close'].round(3))

load_futures('GC')
```

```python
def load_fx(to_symb, from_symb):
    end_date = datetime.now().date()
    start_date = end_date - relativedelta(months= 1)
    start_date = start_date.strftime('%Y-%m-%d')
    df = openbb.forex.load(from_symbol=from_symb, to_symbol=to_symb, start_date=start_date, interval='1day')
    df = df.iloc[[0, -1]]
    return (df.iloc[0]['Adj Close'], df.iloc[1]['Adj Close'])
```

```python
'''
U.S. inflation and labor-market data is the goal
Look at gold futures along side us inflation and labor-market data
Gain insight into this relation ship 
'''


# gold and oil futures data from from Jan 1st 2000 to now 
data = pd.DataFrame(
    openbb.futures.historical(
        symbols= gold['Ticker'],
        start_date= "2000-01-01",
        end_date=str(datetime.now().date())
        )['Close']
).rename(columns={'GC': 'CME Gold Futures', 'QO': 'E-mini Gold'}).reset_index()

data.head()
```

```python
# line plots showing price comparison
fig = go.Figure([
    go.Scatter(
        name='SGU Futures',
        x=data['Date'],
        y=data['SGU'],
        mode='lines',
        line=dict(color='rgb(31, 149, 120)'),
    ),
    go.Scatter(
        name='CME Gold Futures',
        x=data['Date'],
        y=data['CME Gold Futures'],
        mode='lines',
        line=dict(color='rgb(31, 76, 180)')
    ),
    go.Scatter(
        name='E-mini Gold',
        x=data['Date'],
        y=data['E-mini Gold'],
        mode='lines',
        line=dict(color='rgb(131, 25, 180)')
    ),
        go.Scatter(
        name='GCT Gold',
        x=data['Date'],
        y=data['GCT'],
        mode='lines',
        line=dict(color='rgb(131, 25, 10)')
    ),
        go.Scatter(
        name='MGC Gold',
        x=data['Date'],
        y=data['MGC'],
        mode='lines',
        line=dict(color='rgb(31, 125, 180)')
    )
])
fig.update_layout(
    yaxis_title=' $ Price per oz',
    title='Futures Comparison',
    hovermode="x"
)
fig.show()
```
